```python
from os import getenv
from dotenv import load_dotenv

load_dotenv()
AI_BASE_URL = getenv("AI_URL")
API_KEY = getenv("OPENROUTER_API_KEY")
TEST_MODEL = "mistralai/devstral-2512:free"
TEST_MODEL_PROVIDER = "mistralai"


LANGSMITH_TRACING=getenv("LANGSMITH_TRACING")
LANGSMITH_ENDPOINT=getenv("LANGSMITH_ENDPOINT")
LANGSMITH_API_KEY=getenv("LANGSMITH_API_KEY")

print("AI_BASE_URL: ", AI_BASE_URL)
print("API_KEY: ", API_KEY)
print("TEST_MODEL: ", TEST_MODEL)
print("TEST_MODEL_PROVIDER: ", TEST_MODEL_PROVIDER)
print("LANGSMITH_TRACING: ", LANGSMITH_TRACING)
print("LANGSMITH_ENDPOINT: ", LANGSMITH_ENDPOINT)
print("LANGSMITH_API_KEY: ", LANGSMITH_API_KEY)
```

```python
from langchain.chat_models import init_chat_model
# Initialize the model with OpenRouter's base URL
model = init_chat_model(
    model=TEST_MODEL,
    model_provider=TEST_MODEL_PROVIDER,
    base_url=AI_BASE_URL,
    api_key=API_KEY,
    # default_headers={
    #     "HTTP-Referer": getenv("YOUR_SITE_URL"),  # Optional. Site URL for rankings on openrouter.ai.
    #     "X-Title": getenv("YOUR_SITE_NAME"),  # Optional. Site title for rankings on openrouter.ai.
    # }
)

# Example usage
response = model.invoke("你好！")
print(response.content)
```

```python
response = model.invoke("你是谁！")
print(response.content)
```

```python
from langchain.agents import create_agent


def get_weather(city: str) -> str:
    """Get weather for a given city."""
    return f"It's always sunny in {city}!"

agent = create_agent(
    model=model,
    # model_provider=TEST_MODEL_PROVIDER,
    # base_url=AI_BASE_URL,
    
    # api_key=API_KEY,
    tools=[get_weather],
    system_prompt="You are a helpful assistant",
)

# Run the agent
response = agent.invoke(
    {"messages": [{"role": "user", "content": "What is the weather in San Francisco?"}]}
)
# 访问最后一条消息的内容
print(response["messages"][-1].content)
```
